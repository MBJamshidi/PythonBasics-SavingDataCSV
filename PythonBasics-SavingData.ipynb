```python
#Python Basics
#Saving Data
#When we have a dataframe we can work with the data and save the results in other formats

#pandas has the method unique to determine the unique elements in a column of a dataframe

#Ex
#determining the unique year of the albums in the data set
import pandas as pan


ALBUMROCKS={'Years of Publication':[2014, 2009, 1920, 2011, 2009, 2020, 2014], 
            'Name of Album':['Kermanshah', 'Bistoon', 'Iran', 'TOLO', 'Farda', 'Neda', 'Rostam' ],
            'Time':[4.5, 6, 3.5, 8, 9, 12, 16], 
            'Country':['Iran', 'Czech Republic', 'Germany','the United States', 'Australia', \
                       'Canada', 'New Zealand'], 'Num. Members':[12, 3, 6, 19, 31, 2, 7]}
#Defining the dictionary 'ALBUMROCKS' as a dataframe 'DDataFF'
DDataFF=pan.DataFrame(ALBUMROCKS)
print(DDataFF)

#We enter the name of the dataframe 'DDataFF', then enter the name of the column ‘YearsOfPublication’ within brackets
print (DDataFF['Years of Publication'])

#Then we apply the method unique.
print (DDataFF['Years of Publication'].unique())
#The result is all of the unique elements in the column 'Years of Publication'

# Creating a new database consisting of songs from the 2011 and after.
# We can look at the column ‘Years of Publication’ for songs made after 2011
# then select the corresponding rows
#We can accomplish this within one line of code in Pandas:

# 1: We can use the inequality operators for the entire dataframe in pandas
# The result is a series of Boolean values
# For our case, we simply specify the column 'Years of Publication' and the inequality for the albums after 1979
#The result is a series of Boolean values
print(DDataFF['Years of Publication']>=2011)



# The result is true when the condition is true and false otherwise
# We can select the specified columns in one line; we simply use the dataframe’s names,
# and in square brackets we place the previously mentioned inequality and assign it to the variable NewDDataFF
NewDDataFF=DDataFF[DDataFF['Years of Publication']>=2011]
print(NewDDataFF)
# We now have a new dataframe, where each album was released after 2011


#SAVE as CSV
# "to_csv" method
#We can save the new dataframe using the method 'to_csv'
NewDDataFF.to_csv("Name_of_the File.csv")
#The argument is the name of the csv file
#Make sure you include a dot csv extension
#There are other functions to save the dataframe in other formats
```
